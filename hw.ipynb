```python
#Q1:Which airline had the highest percentage of delayed or cancelled flights in 2015?


del_airline = flights[(flights['CANCELLED'] == 1) | (flights['DIVERTED'] == 1) & (flights['YEAR'] == 2015)]['AIRLINE'].value_counts()[:1]
print(del_airline)
del_fl = flights.AIRLINE.value_counts()[0:1]
print(del_fl)
percentage = del_airline/del_fl*100
print(percentage)
```

```python
#Q2:Which airports had the most flight cancellations?
flights.CANCELLED.value_counts()
can_fl = flights[flights['CANCELLED'] == 1].groupby('ORIGIN_AIRPORT').size().idxmax()
print(can_fl)
airports.loc[airports.IATA_CODE == 'ORD']

#QUESTION: HOW TO COUNT ORD CANCELLED FLIGHTS?
```

```python
#Q3:Are there any geographical patterns in flight delays? Do certain 
#airports experience more delays than others?


grouped = flights.groupby('ORIGIN_AIRPORT')['ARRIVAL_DELAY'].sum().reset_index()
sorted_values = grouped.sort_values('ARRIVAL_DELAY', ascending=False)[0:5]
print(sorted_values)
new_sorted_values = sorted_values.rename(columns = ({"ORIGIN_AIRPORT":"IATA_CODE"}))
merged = new_sorted_values.merge(airports,how="inner",on="IATA_CODE")
print("The resultant dataframe is:")
print(merged)

 
```

```python
#Q4:What was the average flight delay for each day of the week? 
#Are weekends or weekdays more prone to delays?

days = flights.groupby('DAY_OF_WEEK')['ARRIVAL_DELAY'].mean()
print("Answer is: weekdays more prone to delays", days)

#QUESTION:how to get it in descending order??
```
